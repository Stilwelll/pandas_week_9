# Homework Excersise 1 - White Sox Data
Return a dataframe showing the White Sox qualified hitters* in order from best to worst XBHA**.

*A hitter must have at least 100 At Bats (AB's) to qualify

** XBHA is calculated as the number of extra base hits (2B, 3B, HR) divided by the number of At Bats (AB's)

```python
import pandas as panda
```

```python
sox_data = panda.read_csv("whitesox_2021_hitting.csv")


sox_data['XBH'] = sox_data['2B'] + sox_data['3B'] + sox_data['HR']

sox_data['XBHA'] = sox_data['XBH'] / sox_data['AB']

sox_data
```

```python
qual_hitters = sox_data[sox_data['AB'] >= 100]
correct_order = qual_hitters.sort_values('XBHA',ascending=True).reset_index(drop=True)
correct_order[['Pos','Rk','Name','XBHA','XBH','AB']]
```
